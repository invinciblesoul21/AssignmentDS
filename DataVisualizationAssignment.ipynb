```python
#Q5 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.


np.random.seed(0)


num_samples = 100


feature1 = np.random.rand(num_samples)
feature2 = feature1 + np.random.normal(scale=0.2, size=num_samples)
feature3 = 0.5 * feature1 + 0.5 * feature2 + np.random.normal(scale=0.1, size=num_samples)


data = pd.DataFrame({'Feature1': feature1, 'Feature2': feature2, 'Feature3': feature3})

correlation_matrix = data.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#BOKEH ASSIGNMENT

#Q1 .Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, legend_label="Sine Wave", line_width=2)

show(p)
```

```python
#Q2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Prepare data
num_points = 100  # Number of data points
x = np.random.rand(num_points) * 10  # Random x values between 0 and 10
y = np.random.rand(num_points) * 10  # Random y values between 0 and 10
sizes = np.random.randint(5, 20, num_points)  # Random sizes between 5 and 20
colors = np.random.choice(['red', 'green', 'blue'], num_points)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh figure
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add a scatter glyph to the figure
p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

# Display the plot
show(p)
```

```python
#Q3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Fruit data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Pears']
counts = [25, 30, 15, 20, 10]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a Bokeh figure
p = figure(x_range=fruits,  # Set x-axis to categorical fruit names
           title="Fruit Counts",
           x_axis_label="Fruit",
           y_axis_label="Count")

# Add a bar glyph
p.vbar(x='fruits', top='counts', width=0.9, source=source)

# Customize the plot
p.xgrid.grid_line_color = None  # Remove vertical grid lines
p.y_range.start = 0  # Start y-axis at 0

# Display the plot
show(p)
```

```python
#Q4 Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Data: Assuming you have 'hist' and 'edges' already calculated
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(top=hist,
                                    bottom=np.zeros(len(hist)),
                                    left=edges[:-1],
                                    right=edges[1:]))

# Create a Bokeh figure
p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")

# Add a quad glyph for the histogram bars
p.quad(top='top', bottom='bottom', left='left', right='right',
       source=source, fill_color="skyblue", line_color="white")

# Display the plot
show(p)
```

```python
#Q5 Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar

# Data: Assuming you have 'data_heatmap', 'xx', and 'yy' already defined
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)

# Reshape data for ColumnDataSource
data_source = ColumnDataSource(data=dict(x=xx.flatten(),
                                         y=yy.flatten(),
                                         colors=data_heatmap.flatten()))

# Create color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create Bokeh figure
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()))

# Add heatmap rectangles
p.rect(x='x', y='y', width=x[1]-x[0], height=y[1]-y[0],
       source=data_source, fill_color={'field': 'colors', 'transform': color_mapper},
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12,
                     border_line_color=None, location=(0,0))
p.add_layout(color_bar, 'right')

# Display the plot
show(p)
```
