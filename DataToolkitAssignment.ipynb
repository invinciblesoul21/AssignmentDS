```python
#Q3 Explain the following terms:
#1 The difference in np.array, np.asarray and np.asanyarray.
#2 The difference between Deep copy and shallow copy.

#1 np.array: Always creates a copy, even if the input is already an ndarray.
#np.asarray: Creates a copy only if necessary (e.g., if the input is not an ndarray). It does not preserve subclasses.
#np.asanyarray: Avoids creating a copy if possible and preserves the subtype if the input is an ndarray subclass.

#2 Shallow copy is in which change occurs in both original and derived array whereas in deep copy only derived array got changed not the original one.
```
